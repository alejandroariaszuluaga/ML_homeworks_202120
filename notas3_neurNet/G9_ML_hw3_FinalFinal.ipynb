### Funciones Auxiliares

A continuación se crean algunas funciones que serán utilizadas más adelante.

Función `calc_iv` para obtener las variables IV (information value) y WoE (weight of evidence). Estas variables, de forma general, nos permiten conocer la importancia de cada feature disponible.

Puede encontrar más información en:
- https://www.kaggle.com/puremath86/iv-woe-starter-for-python
- https://www.listendata.com/2015/03/weight-of-evidence-woe-and-information.html

````python
# Cálculo de IV
def calc_iv(df, feature, target, pr=False):
    lst = []
    df[feature] = df[feature].fillna("NULL")

    for i in range(df[feature].nunique()):
        val = list(df[feature].unique())[i]
        lst.append([feature,                                                        # Variable
                    val,                                                            # Value
                    df[df[feature] == val].count()[feature],                        # All
                    df[(df[feature] == val) & (df[target] == 0)].count()[feature],  # Good (think: Fraud == 0)
                    df[(df[feature] == val) & (df[target] == 1)].count()[feature]]) # Bad (think: Fraud == 1)

    data = pd.DataFrame(lst, columns=['Variable', 'Value', 'All', 'Good', 'Bad'])
    data['Share'] = data['All'] / data['All'].sum()
    data['Bad Rate'] = data['Bad'] / data['All']
    data['Distribution Good'] = (data['All'] - data['Bad']) / (data['All'].sum() - data['Bad'].sum())
    data['Distribution Bad'] = data['Bad'] / data['Bad'].sum()
    data['WoE'] = np.log(data['Distribution Good'] / data['Distribution Bad'])
    
    data = data.replace({'WoE': {np.inf: 0, -np.inf: 0}})

    data['IV'] = data['WoE'] * (data['Distribution Good'] - data['Distribution Bad'])

    data = data.sort_values(by=['Variable', 'Value'], ascending=[True, True])
    data.index = range(len(data.index))

    if pr:
        print(data)
        print('IV = ', data['IV'].sum())

    iv = data['IV'].sum()
    print('El IV de esta variable es:',iv)
    print(df[feature].value_counts())
    return iv, data
````

````python
# Codificación One-Hot
def convert_dummy(df, feature,rank=0):
    pos = pd.get_dummies(df[feature], prefix=feature)
    mode = df[feature].value_counts().index[rank]
    biggest = feature + '_' + str(mode)
    pos.drop([biggest],axis=1,inplace=True)
    df.drop([feature],axis=1,inplace=True)
    df=df.join(pos)
    return df
````

````python
def get_category(df, col, binsnum, labels, qcut = False):
    if qcut:
        localdf = pd.qcut(df[col], q = binsnum, labels = labels) # quantile cut
    else:
        localdf = pd.cut(df[col], bins = binsnum, labels = labels) # equal-length cut
        
    localdf = pd.DataFrame(localdf)
    name = 'gp' + '_' + col
    localdf[name] = localdf[col]
    df = df.join(localdf[name])
    df[name] = df[name].astype(object)
    return df
````

````python
# Matriz de Confusión
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        
    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Etiqueta')
    plt.xlabel('Predicción')
````

````python
#Funciones de activación y sus derivadas
def relu(x):
    return np.maximum(0,x)

def sigma(x):
    res_sigmoid = 1 / (1 + np.exp(-(x))) # Salida de la función sigmoide #
    return res_sigmoid

def sigma_deriv(x):
    deriv = sigma(x) * (1-sigma(x))
    return deriv

def relu_deriv(x):
    if x > 0:
        relu_der = 1
    else:
        relu_der = 0
    return relu_der
````

````python
#Paso Forward
def calculate_forward(weights, xActual, layers, activations): 
    activations.append(xActual)
    for j in range (len(hidden_layers)):     #Capas escondidas
        z=[]
        for i in range (1,hidden_layers[j]+1):
            a = np.dot(activations[j], weights[j][i,:]) # se calcula el valor de a
            z.append(relu(a)) # Se aplica la función de activación, se listan los valores de z calculados
        activations.append(np.transpose(z))  # se agregan los valores de z en la lista activations
        activations[j+1]=np.insert(activations[j+1],0,1) #se agrega 1 como término del bias
````

````python
#Neurona de salida
def calculate_output(weights, activations):
    a_out = np.dot(activations[len(activations)-1], weights[len(activations)-1][0,:])
    z_out = sigma(a_out)
    return (z_out, a_out)
````

````python
#Backpropagation delta_out
def calculate_delta_out(yActual, z_out, a_out):
    delta_out = (z_out-yActual)
    return delta_out
````

````python
#Backpropagation cálculo de los deltas j
def calculate_delta_j(weights, activations, delta_out, hidden_layers):
    delta_j=[]
    delta_j2=[]
    delta_j1=[]

    for j in reversed(range(1,len(activations))):
        if j == len(activations)-1: # Deltas en la capa última capa escondida
            for i in range(hidden_layers[j-1]+1):
                d_j = relu_deriv(activations[j][i])*np.dot(weights[j][0,i], delta_out)
                delta_j2.append(d_j)
            delta_j.append(np.transpose(delta_j2))
        else:
            for i in range(hidden_layers[j-1]+1): #Deltas en las demás capas escondidas
                d_j = relu_deriv(activations[j][i])*np.dot(delta_j,weights[j][i,:])
                delta_j1.append(d_j)
            delta_j.append(np.transpose(delta_j1))
    delta_j[-1]=delta_j[-1][0] #Se organiza el valor como una lista
    delta_j.reverse() #Se reordenan los datos
    delta_j.append([delta_out]) #Se agrega el delta_out al arreglo
    return delta_j
````

````python
#Actualización de todos los pesos
def update_weights(weights, delta_j, activations, n):
    for i in range (len(layers)-1): #pesos en la capa i
        for j in range(layers[i+1]): # deltas en la neurona j de la capa i
            for k in range(layers[i]): # activaciones k (zetas)
                w = weights[i][j,k]-n*activations[i][k]*delta_j[i][j] # Actualización de cada peso
                weights[i][j,k] = w
````
