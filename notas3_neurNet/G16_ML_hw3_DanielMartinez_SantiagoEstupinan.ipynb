### Funciones Auxiliares

A continuación se crean algunas funciones que serán utilizadas más adelante.

Función `calc_iv` para obtener las variables IV (information value) y WoE (weight of evidence). Estas variables, de forma general, nos permiten conocer la importancia de cada feature disponible.

Puede encontrar más información en:
- https://www.kaggle.com/puremath86/iv-woe-starter-for-python
- https://www.listendata.com/2015/03/weight-of-evidence-woe-and-information.html

````python
# Cálculo de IV
def calc_iv(df, feature, target, pr=False):
    lst = []
    df[feature] = df[feature].fillna("NULL")

    for i in range(df[feature].nunique()):
        val = list(df[feature].unique())[i]
        lst.append([feature,                                                        # Variable
                    val,                                                            # Value
                    df[df[feature] == val].count()[feature],                        # All
                    df[(df[feature] == val) & (df[target] == 0)].count()[feature],  # Good (think: Fraud == 0)
                    df[(df[feature] == val) & (df[target] == 1)].count()[feature]]) # Bad (think: Fraud == 1)

    data = pd.DataFrame(lst, columns=['Variable', 'Value', 'All', 'Good', 'Bad'])
    data['Share'] = data['All'] / data['All'].sum()
    data['Bad Rate'] = data['Bad'] / data['All']
    data['Distribution Good'] = (data['All'] - data['Bad']) / (data['All'].sum() - data['Bad'].sum())
    data['Distribution Bad'] = data['Bad'] / data['Bad'].sum()
    data['WoE'] = np.log(data['Distribution Good'] / data['Distribution Bad'])
    
    data = data.replace({'WoE': {np.inf: 0, -np.inf: 0}})

    data['IV'] = data['WoE'] * (data['Distribution Good'] - data['Distribution Bad'])

    data = data.sort_values(by=['Variable', 'Value'], ascending=[True, True])
    data.index = range(len(data.index))

    if pr:
        print(data)
        print('IV = ', data['IV'].sum())

    iv = data['IV'].sum()
    print('El IV de esta variable es:',iv)
    print(df[feature].value_counts())
    return iv, data
````

````python
# Codificación One-Hot
def convert_dummy(df, feature,rank=0):
    pos = pd.get_dummies(df[feature], prefix=feature)
    mode = df[feature].value_counts().index[rank]
    biggest = feature + '_' + str(mode)
    pos.drop([biggest],axis=1,inplace=True)
    df.drop([feature],axis=1,inplace=True)
    df=df.join(pos)
    return df
````

````python
def get_category(df, col, binsnum, labels, qcut = False):
    if qcut:
        localdf = pd.qcut(df[col], q = binsnum, labels = labels) # quantile cut
    else:
        localdf = pd.cut(df[col], bins = binsnum, labels = labels) # equal-length cut
        
    localdf = pd.DataFrame(localdf)
    name = 'gp' + '_' + col
    localdf[name] = localdf[col]
    df = df.join(localdf[name])
    df[name] = df[name].astype(object)
    return df
````

````python
# Matriz de Confusión
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        
    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Etiqueta')
    plt.xlabel('Predicción')
````

## Regresión Logística

Inicialmente se probará un modelo de regresión logística para tener una referencia (también se conoce como _baseline_) y comprobar que un modelo de red neuronal permite obtener mejores resultados.

````python
#Backpropagation 5 neuronas pocas capas
import random

# Se define la función de activación RELU
def RELU(a):
    if a < 0:
        return 0
    else:
        return a
    
# Se define la derivada de la función de activación RELU
def DERELU(a):
    if a<0:
        return 0
    else:
        return 1
    
errores = [] # Se define una lista para guardar el progreso del error del modelo.

# Se define la función de entrenamiento del modelo, donde se realizará el backpropagation.
# La red neuronal definida puede cambiar sus neuronas por capa y el número de iteraciones, pero tiene fijo el # de capas
def backpropagation(x,y,nneur, niters):
    w01 = np.random.rand(x.shape[1],nneur) # Se inicializan los pesos desde la entrada a la capa 1.
    w1  = w01.copy()
    w02 = np.random.rand(nneur,nneur) # Se inicializan los pesos desde la capa 1 a la capa 2.
    w2  = w02.copy()
    w03 = np.random.rand(nneur,1) # Se inicializan los pesos desde la capa 2 a la capa de salida.
    w3  = w03.copy()
    z1  = np.array([]) # Se inicializa la lista donde se guardarán los zs de la capa 1.
    z2  = [] # Se inicializa la lista donde se guardarán los zs de la capa 2.
    z0  = [] # Se inicializa la lista donde se guardarán los zs de la capa de salida.
    LR  = 0.0001 # Se inicializa el learning rate.
    Der0= [] # Se inicializa la lista donde se guardarán las derivadas de la funcion de activación de capa 1.
    Der1= [] # Se inicializa la lista donde se guardarán las derivadas de la funcion de activación de capa 2.

    i=random.randrange(0, x.shape[0]-1) # Se toman valores iniciales aleatorios del conjunto de prueba.
    x0=x[i]
    y0=y[i]
    
    # Se comienza a iterar en el algoritmo
    for i in range(niters):
        #Cálculo de zj
        
        a=np.dot(x0,w1) # Se calculan las a de primera capa.
        for g in a:
            Der0.append(DERELU(g)) # Se calculan las derivadas de RELU de primera capa.
        for j in a:
            z1 = np.append(z1, RELU(j)) # Se calculan las z de primera capa.
        a1=np.dot(z1,w2) # Se calculan las a de segunda capa.
        
        for f in a1:
            Der1.append(DERELU(f)) # Se calculan las derivadas de RELU de segunda capa.
        for k in a1:
            z2.append(RELU(k)) # Se calculan las z de segunda capa.
           
        z0=np.dot(z2,w3) # Se calcula z de capa de salida.
        
        ### COMIENZA BACKPROP
        
        d0=z0-y0 # Se calcula el error (delta 0 - delta capa de salida)
        errores.append(d0) # Se agregan los errores a la lista de errores para graficar la disminución del error
        
        d2=Der1*np.dot(w3,d0) # Se calcula el delta de segunda capa
        
        d1=Der0*np.dot(w2,d2) # Se calcula el delta de primera capa
        
        
        g3=d0*z2 # Se calculan los gradientes para las conexiones de segunda capa-capa de salida
        g2=d2*(z1[:,None]) # Se calculan los gradientes para las conexiones de primera capa-segunda capa
        g1=x0*(d1[:,None]) #Se calculan los gradientes para las conexiones de entradas-primera capa
      
        # Se actualizan los pesos de las conexiones usando learninf rate y los gradientes.
        subs = LR*g3
        w3=np.subtract(w3, subs[:,None])
        w2=w2-LR*g2
        w1=w1-np.transpose(LR*g1)
        
        # Se resetean las variables
        z1  = np.array([])
        z2  = []
        z0  = []
        Der0= []
        Der1= []
        
        # Se elige un nuevo punto al azar.
        i=random.randrange(0, x.shape[0]-1)
        x0=x[i]
        y0=y[i] 
    return w1, w2, w3
        
            
X_train_array = np.c_[np.ones(X_train.shape[0]), X_train]
y_train_array = y_train.values     
    
        
w1, w2, w3 = backpropagation(X_train_array, y_train_array, 5, 500) 

````
