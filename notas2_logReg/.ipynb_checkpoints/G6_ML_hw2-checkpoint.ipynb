```python
x = heart_df.values
scaler = MinMaxScaler() # Utilice un MinMaxScaler en esta ocasión #
x_scaled = scaler.fit_transform(x) # Aplique el escalamiento #
heart_df_scaled = pd.DataFrame(x_scaled, columns=heart_df.columns) # Conversión a un DataFrame #
```

```python
x = heart_df_scaled.iloc[:,:-1]
y = heart_df_scaled.iloc[:,-1]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

## Parte 1

A continuación, debe utilizar la función `LogisticRegression` de SciKit-Learn para obtener un primer modelo de regresión logística y calcular su precisión a partir de la función `accuracy_score`.

Obtenga resultados sobre los datos de entrenamiento y los datos de prueba. Concluya al respecto.

```python
logreg = LogisticRegression() # Modelo de Regresión Logística #
logreg.fit(x_train, y_train) # Entrenamiento sobre los Datos #
y_pred = logreg.predict(x_test) # Predicción de Etiquetas #
prec = accuracy_score(y_test, y_pred)
print(f'Precisión inicial: {prec}')
```

Como se puede observar, se obtuvo una precisión de 0.8453, lo que quiere decir que el modelo de regresión logística fue capáz de clasificar correctamente el $84.53\%$ de las muestras de prueba.

### Matriz de Confusión

Como recordará, la información en este dataset no posee una distribución de clases balanceada. La matriz de confusión es una herramienta que permite analizar el comportamiento del modelo para estos datos.

Puede encontrar más información en este artículo: [Understanding Confusion Matrix](https://towardsdatascience.com/understanding-confusion-matrix-a9ad42dcfd62)

```python
cm = confusion_matrix(y_test,y_pred)
conf_matrix = pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
plt.figure(figsize = (8,5))
sn.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu")
```

Para poder analizar el resultado de la matriz de confusión, se debe entender que esta se divide en cuatro cuadrantes: *verdadero positivo*, *falso positivo* (error tipo I), *falso negativo* (error tipo II)y *verdadero negativo*.

* **Verdadero positivo**: Se predijo positivo (1) y es verdad
* **Verdadero negativo**: Se predijo negativo (0) y es verdad
* **Falso positivo**: Se predijo positivo (1) y es falso
* **Falso negativo**: Se predijo negativo (0) y es falso

Con esto en mente, se puede observar que, para el conjunto de datos en cuestión, se obtuvieron 624 verdaderos negativos, 10 verdaderos positivos, 112 falsos negativos y 4 falsos positivos. Como se puede observar, la cantidad de errores tipo I es muy grande en comparación con la cantidad de errores tipo II (se cometen 24 veces más errores tipo I que tipo II), y esto es muy problemático en el contexto en que se trabaja; se prefiere predecir a un paciente con riesgo de enfermedad coronaria y que al final no la padezca, a que se prediga un paciente sin riesgo de la enfermedad y que al final la padezca. Por tal motivo, en la siguiente sección se trabajara en el tratamiendo de este desbalance

### Tratamiento de Datos Desbalanceados

En casos de estudio como identificación de tendencias patológicas, es común tener datos desbalanceados, o en los cuales la gravedad de cometer [errores tipo I o tipo II](https://www.simplypsychology.org/type_I_and_type_II_errors.html) son bastante diferentes. En muchos casos, se puede preferir cometer cierto tipo de error sobre el otro. Para trabajar con estos problemas, se suele preferir sacrificar algo de [precisión](https://developers.google.com/machine-learning/crash-course/classification/accuracy) del modelo, a cambio de una mejora en términos prácticos.

Este proceso se realiza a partir de la modificación del umbral a partir del cual se decide si un dato pertenece a una clase u otra. Utilice la función `binarize` para obtener valores de `0` ó `1` según las probabilidades de las predicciones realizadas utilizando la función `model.predict_proba`.

Algunas métricas utilizadas para interpretar el rendimiento del modelo son las siguientes:

- F1Score
- TP Rate / Recall / Sensitividad
- TN Rate / Especificidad

Puede encontrar más información sobre estas métricas en los artículos:
- [Accuracy, Precision, Recall or F1?](https://towardsdatascience.com/accuracy-precision-recall-or-f1-331fb37c5cb9)
- [Sensitivity Vs Specificity In Data Science](https://medium.com/fintechexplained/sensitivity-vs-specificity-in-data-science-2f673039dbd1)

```python
from sklearn.preprocessing import binarize

umbrales = np.linspace(0, 1, 20)# Vector que Contiene los Umbrales a Analizar #

for umbral in umbrales:
    cm2 = 0
    y_pred_prob_yes = logreg.predict_proba(x_test)# Utilice la función predict_proba para calcular las probabilidades de pertenecer a la clase 1 #
    y_pred2 = binarize(y_pred_prob_yes, threshold = umbral)# Utilice la función binarize para convertir las probabilidades a clases 0 y 1 #
    cm2 = confusion_matrix(y_test,y_pred2[:, 1])
    
    sen = cm2[1,1]/(float(cm2[1,1]+cm2[1,0])) # Cálculo de la sensitividad #
    esp = cm2[0,0]/(float(cm2[0,0]+cm2[0,1])) # Cálculo de la especificidad #
    
    print(f'''->Para un umbral de {round(umbral,4)} tenemos esta matriz de confusión:
{cm2}
Con {cm2[0,0]+cm2[1,1]} predicciones correctas y {cm2[1,0]} falsos negativos.
Sensitividad: {round(sen, 4)} Especificidad: {round(esp, 4)}\n''')
```

En la celda anterior se presentan las matrices de confusión obtenidas por cada umbral utilizado. De igual forma, se presentan junto a cada matriz el parámetro de *sensitividad* y *especificidad*, los cuales vienen definidos mediante:

$$\text{Sensitividad} = \frac{VP}{VP + FN}$$
<br>
$$\text{Especificidad} = \frac{VN}{VN + FP}$$

El primer parámetro expresa que tan bueno es el modelo para predecir VP, y el segundo expresa que tan bueno es el modelo para predecir verdaderos negativos. Para el caso de diagnostico en medicina, se prefiere el poder acertar con la cantidad de VN, pero de igual forma se desea disminuir la cantidad de FN. Por esto, se prefiere llegar a un balance entre FP con FN.

En este caso, se puede observar que el umbral que presenta mejores resultados es el umbral de 0.2632, generando una sensitividad de 0.3361 y una especificidad de 0.8838

Por otra parte, un criterio que facilita el visualizar este trade-off se conoce como la Curva ROC, en donde se ubican múltiples puntos para distintos valores del umbral de clasificación. Puede leer el artículo a continuación para una explicación más detallada:
- Curva ROC: [Understanding AUC - ROC Curve](https://towardsdatascience.com/understanding-auc-roc-curve-68b2303cc9c5)

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob_yes[:,1])
plt.figure(figsize = (10, 5))
plt.plot(fpr,tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('Curva ROC para Clasificador de Enfermedad Coronaria')
plt.xlabel('Tasa de Falsos Positivos (1-Especificidad)')
plt.ylabel('Tasa de Verdaderos Positivos (Sensitividad)')
plt.grid(True)
```

Observe la gráfica, y teniendo en cuenta los resultados concluya sobre la proporción que se puede considerar más adecuada en este contexto.

De acuerdo a la curva ROC, el umbral que mejor aporta al modelo de clasificación es aquel en donde la coordenada $(\text{fpr}_{i}, \text{tpr}_{i})$ esta cercana a la coordenada $(0, 1)$. Entonces, para poder saber cuál es el mejor umbral para este contexto, se procede primero a tomar cada par $(\text{fpr}_{i}, \text{tpr}_{i})$ y calcular su distancia con la esquina superior izquierda de la figura:

$$\text{dist}_{i} = \sqrt{(\text{fpr}_{i})^{2} + (\text{tpr}_{i} - 1)^{2}}$$

```python
dist = []
for i in range(len(fpr)):
    dist.append((fpr[i]**2) + ((tpr[i] - 1)**2)**0.5)
dist = np.asarray(dist)

i = np.argmin(dist)
print(f'El umbral adecuado en este contexto es {round(thresholds[i], 4)}')
```

## Parte 2

Ahora usted debe desarrollar su propia implementación del método Descenso de Gradiente estocástico para un modelo de regresión logística. Realice el entrenamiento a partir del 80% de los datos. Al final del entrenamiento, en la lista errores se deben tener los valores de la función de error para cada iteración y así poder observar el progreso gráficamente.

```python
np.random.seed(22)
X_train_array = np.c_[np.ones(x_train.values.shape[0]), x_train.values]
y_train_array = y_train.values

w0 = np.random.rand(X_train_array.shape[1]) # Inicialización de w #

# Inicialización de criterio de parada (al disminuir el valor de eps, se aumenta la precisión y el tiempo de procesamiento)
eps = 1e-5
error = 100
dif = 100

def sigma(w,x):
    z = np.dot(w, x) # Se realiza el priducto interno entre w y x #
    res_sigmoid = 1/(1 + np.exp(-z)) # Salida de la función sigmoide #
    return res_sigmoid

tasas = np.linspace(0, 0.1, 10) # Defina un vector que contenga las tasas que desea probar #
for tasa in tasas:
    errores = [error] # Agregar a esta lista los valores de la función de error en cada iteración #
#     while error>eps:         # Criterio de parada 1 #
    for j in range (0, 1000):  # Criterio de parada 2 #
        # # Desarrollo de algoritmo # #
        i = np.random.randint(X_train_array.shape[0]) # Posición aleatoria #
        xActual = X_train_array[i]
        yActual = y_train_array[i]

        g = sigma(w0, xActual) #Se evalua el producto interno entre w y x en la función sigmoide#
        e = g - yActual #Se calcula el error#
        errores.append(e)

        w = w0 - (tasa*e*xActual) #Se actualiza el vector de pesos#
        w0 = w.copy()
        # # # # # # # # # # # # # # # #
    
    X_test_array = np.c_[np.ones(x_test.values.shape[0]), x_test.values]
    y_test_array = y_test.values

    y_pred = np.zeros(len(x_test))

    for i in range(0,len(x_test)):
      prob = sigma(w,X_test_array[i])
      if prob>0.5:
        y_pred[i] = 1
      else:
        y_pred[i] = 0

    print(f'Precisión modelo actual: {sklearn.metrics.accuracy_score(y_test,y_pred)}')
    plt.figure(figsize = (15,5))
    plt.plot(errores, color = 'b')
    plt.xlabel('# Iteraciones')
    plt.ylabel('Función de Error')
    plt.title('Descenso de Gradiente Estocástico con ' + r'$\eta = $' + f'{tasa}')
    plt.grid()
    plt.show()
```

Como se puede observar, no se necesitó una tasa grande para que el modelo llegara a un accuracy similar al obtenido mediante la librería utilizada en la parte 1. La mejor precisión obtenida fue $0.84\bar{6}$ con una tasa de $\eta = 0.0\bar{8}$. Para el algoritmo, se utilizó un numero predeterminado de iteraciones ya que, al intentar utilizar el criterio de parada $|\mathrm{w}_{0} - \mathrm{w}| > \epsilon$, no se actualizaba el vector de pesos por alguna razón.
